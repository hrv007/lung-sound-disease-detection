```python
import librosa
import librosa.display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ICBHI 2017/Respiratory_Sound_Database/Respiratory_Sound_Database/patient_diagnosis.csv")
df.head()
```

```python
sr_no = {'101':'URTI'}
for i, j in zip(df['101'].unique(), df['URTI']):
    sr_no[str(i)] = j
```

```python
sr_no.keys()
```

```python
import os
sound_files = os.listdir('ICBHI 2017/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files')
```

```python
req_file_names = []

for i in sound_files:
    if i[-3:]=='wav':
        req_file_names.append([i])
```

```python
labels = []
for i in range(len(req_file_names)):
    req_file_names[i].append(sr_no[req_file_names[i][0][:3]])
    labels.append(sr_no[req_file_names[i][0][:3]])
```

```python
os.mkdir('Chromagrams')
os.mkdir('Chromagrams/Original')
os.mkdir('Chromagrams/Time Stretch')
os.mkdir('Chromagrams/Audio Shift')
os.mkdir('Chromagrams/Pitch Shift')
```

```python
for i in range(len(req_file_names)):

    data, icbhi_sr = librosa.load('Butterworth Filter/'+req_file_names[i][0])

    rb_chroma = librosa.feature.chroma_stft(data)

    plt.figure(figsize=(15,5))
    librosa.display.specshow(rb_chroma)
    plt.xlabel("Time")
    plt.ylabel("Chroma Index")
    plt.colorbar()
    plt.savefig('Chromagrams/Original/'+req_file_names[i][0][:-4]+'.png')
    plt.close()
```

```python
for i in range(len(req_file_names)):

    data, icbhi_sr = librosa.load('Augmented Data/Time Stretch/'+req_file_names[i][0])

    rb_chroma = librosa.feature.chroma_stft(data)

    plt.figure(figsize=(15,5))
    librosa.display.specshow(rb_chroma)
    plt.xlabel("Time")
    plt.ylabel("Chroma Index")
    plt.colorbar()
    plt.savefig('Chromagrams/Time Stretch/'+req_file_names[i][0][:-4]+'.png')
    plt.close()
```

```python
for i in range(len(req_file_names)):

    data, icbhi_sr = librosa.load('Augmented Data/Audio Shift/'+req_file_names[i][0])

    rb_chroma = librosa.feature.chroma_stft(data)

    plt.figure(figsize=(15,5))
    librosa.display.specshow(rb_chroma)
    plt.xlabel("Time")
    plt.ylabel("Chroma Index")
    plt.colorbar()
    plt.savefig('Chromagrams/Audio Shift/'+req_file_names[i][0][:-4]+'.png')
    plt.close()
```

```python
for i in range(len(req_file_names)):

    data, icbhi_sr = librosa.load('Augmented Data/Pitch Shift/'+req_file_names[i][0])

    rb_chroma = librosa.feature.chroma_stft(data)

    plt.figure(figsize=(15,5))
    librosa.display.specshow(rb_chroma)
    plt.xlabel("Time")
    plt.ylabel("Chroma Index")
    plt.colorbar()
    plt.savefig('Chromagrams/Pitch Shift/'+req_file_names[i][0][:-4]+'.png')
    plt.close()
```
