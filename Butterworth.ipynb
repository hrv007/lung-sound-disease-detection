# Meditron

```python
def butter_bandpass(lowcut, highcut, fs, order=5):
    return butter(order, [lowcut, highcut], fs=fs, btype='band')

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y

if __name__ == "__main__":
  # Generate some noisy data.
    
    data, icbhi_sr = librosa.load('ICBHI 2017/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files/101_1b1_Pr_sc_Meditron.wav')
    
    filtered_data = butter_bandpass_filter(data, 250, 2000, fs=icbhi_sr, order=4)

  # Plot the original and filtered data.
    plt.figure(figsize=(15,8))
    plt.plot(data)
    plt.plot(filtered_data)
    plt.show()
    
    sf.write("New_filter.wav", filtered_data, icbhi_sr)
```

```python
plt.figure(figsize=(15, 8))
plt.plot(filtered_data)
plt.show()
```

# Applying filter to all sounds

```python
import pandas as pd

df = pd.read_csv("ICBHI 2017/Respiratory_Sound_Database/Respiratory_Sound_Database/patient_diagnosis.csv")
df.head()
```

```python
sr_no = {'101':'URTI'}
for i, j in zip(df['101'].unique(), df['URTI']):
    sr_no[str(i)] = j
```

```python
sr_no.keys()
```

```python
import os
sound_files = os.listdir('ICBHI 2017/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files')
```

```python
req_file_names = []

for i in sound_files:
    if i[-3:]=='wav':
        req_file_names.append([i])

req_file_names
```

```python
labels = []
for i in range(len(req_file_names)):
    req_file_names[i].append(sr_no[req_file_names[i][0][:3]])
    labels.append(sr_no[req_file_names[i][0][:3]])
```

```python
def butter_bandpass(lowcut, highcut, fs, order=5):
    return butter(order, [lowcut, highcut], fs=fs, btype='band')

def butter_bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y
```

```python
for i in range(len(req_file_names)):

    data, icbhi_sr = librosa.load('ICBHI 2017/Respiratory_Sound_Database/Respiratory_Sound_Database/audio_and_txt_files/'+req_file_names[i][0])
    
    if req_file_names[0][0][-12:-4]=='Meditron':
        order_x = 4
    elif req_file_names[0][0][-12:-4]=='LittC2SE':
        order_x = 4
    elif req_file_names[0][0][-12:-4]=='AKGC417L':
        order_x = 1
    elif req_file_names[0][0][-12:-4]=='Litt3200':
        order_x = 8
    
    filtered_data = butter_bandpass_filter(data, 250, 2000, fs=icbhi_sr, order=order_x)
    
    sf.write("Butterworth Filter/"+req_file_names[i][0], filtered_data, icbhi_sr)
```

# Mel Spectrograms

```python
from librosa.feature import mfcc, melspectrogram
from librosa.display import specshow
import librosa
```

```python
for i in range(len(req_file_names)):

    data, icbhi_sr = librosa.load('Butterworth Filter/'+req_file_names[i][0])

    mel_spec = librosa.feature.melspectrogram(y=data, n_mels=128)

    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

    plt.figure(figsize=(14, 8))
    specshow(mel_spec_db, y_axis='log', fmax=8000, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel spectrogram')
    plt.tight_layout()
    plt.savefig('Mel Spectrogram/'+req_file_names[i][0][:-4]+'.png')
    plt.close()
```
